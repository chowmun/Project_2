#### B.4. Imputing with Median
The values in these features have a wide range, it is more appopriate to use median which is less bias in our case. 
##### B.4i. Mas Vnr Area 

```python
#Impute 'Mas Vnr Area' based on the median of its 'Mas Vnr Type'

median_area = train.pivot_table(index ='Mas Vnr Type', values ='Mas Vnr Area' , aggfunc = 'median')

#define a function that only change the 'Mas Vnr Area' if it is NA
def impute_median_neigh(cols):
    Mas_Vnr_Area = cols[0]
    Mas_Vnr_Type = cols[1]
    
    if pd.isnull(Mas_Vnr_Area):
        return median_area['Mas Vnr Area'][median_area.index == Mas_Vnr_Type]
    else:
        return Mas_Vnr_Area

train['Mas Vnr Area'] = train[['Mas Vnr Area','Mas Vnr Type']].apply(impute_median_neigh,axis=1)
```

#carry out for test data

```python
test['Mas Vnr Area'] = test[['Mas Vnr Area','Mas Vnr Type']].apply(impute_median_neigh,axis=1)
```

#### 2a. Exter Qual and Exterior 1st/2nd
We will carry out Chi-squared test to examine the relationship between Exter Qual and Exterior 1st/2nd.<br>
The chi-square test relies on the difference between observed and expected values.<br>
Our hypothesis for this test will be:<br>
H0 = Exter Qual and Exterior 1st/2nd are independent to each other<br>
H1 = Exter Qual are Exterior 1st/2nd are dependent to each other.

```python
contingency_table=pd.crosstab(train["Exterior 1st"], train['Exter Qual'])
contingency_table
```

```python
Observed_Values = contingency_table.values 
Observed_Values
```

```python
#chi2_contingency: computes the chi-square statistic and p-value for the hypothesis test of independence of the observed frequencies
#in the contingency table
b = scipy.stats.chi2_contingency(contingency_table)
#The output shows the chi-square statistic, the p-value and the degrees of freedom followed by the expected values.
Expected_Values = b[3]
b
```

```python
alpha=0.05 #Find the critical value for 95% confidence
critical_value = chi2.ppf(q = 1-alpha, df=42)
critical_value
```

```python
print('Significance level: ',alpha)
print('Degree of Freedom: ',b[2])
print('chi-square statistic:',b[0])
print('critical_value:',critical_value)
print('p-value:',b[1])
```

chi-square statistic > critical value and p value < alpha, we reject H0.<br>
There is a relationship between 'Exter Qual' and exterior materials, I decided to take only**'Exter Qual'** as one of the feature.
