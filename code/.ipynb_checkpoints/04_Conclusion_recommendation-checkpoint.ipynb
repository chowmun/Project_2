```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
from scipy.stats import chi2
%matplotlib inline
```

```python
pd.options.display.max_columns =999 #dispaly the whole grid

train2_csv = '../datasets/train2.csv'
train2 = pd.read_csv(train2_csv)
```

### Conclusion

1.Based on R2, 85% of the SalePrice is explained by this multi linear model and the predicted sale price is $31K diverted from the real sale price<br>

2.The housing sale price in Ames is increases with the increase in the house total area, garage area, overall house quality and kitchen quality. Certain neighborhood boosts the sale price as well, may be the neighborhood is nearer to some facilities like shopping mall and university.<br>

3.The older a house is, the cheaper its selling price. <br>

4.There are some outliers in the data which our model is not able to handle it well, this could be something to be improved on in the future. 

### Recommendations
1.Recommendation to increase the house sale price:

```python
sns.set(font_scale=1.8)
sns.lmplot(x='years_of_house', y='SalePrice', data=train2, hue='reno', col='Overall Qual', 
           fit_reg=False, size=5, col_wrap=5,scatter_kws={'marker':'o','alpha':0.7,'s':15},palette='Set2')
```

1a. From the model, we know that years of house will reduce the house's value so it is a common practise for house owner to renovate thier house if it is older than 60 years. <br>
From the scatter plots above, it shows that renovation has improved the overall house quality and reduce the impact of years of house. We need to find the balance between the cost of renovation and the increase in the sale price.<br>


```python
train2['SalePrice'].groupby(train2['Kitchen Qual']).median().sort_values().plot(kind='bar',color='blue', figsize=(7,4))
plt.ylabel('SalePrice')
```

1b.From the plot, we can notice that by having a better grade of kitchen, the sale price will be increased. It could be a good idea to improve the kitchen quality through renovation. 

2.Recommendation for neighborhood house with a potential gain

```python
#create a dataframe to get the mean for every neighborhood for each of the year sold.
df=train2.groupby(['Yr Sold','Neighborhood']).mean()
```

```python
#need to reset index as the dataframe treats 'Yr Sold' and 'Neighborhood'
df.reset_index(inplace=True)
```

```python
#plot graphs for the sales price ver year of sold for different neighborhood
sns.set(font_scale=1)
cond_plot = sns.FacetGrid(data=df, col='Neighborhood', col_wrap=5, aspect=1.5)
cond_plot.map(plt.plot, 'Yr Sold' , 'SalePrice')
```

2a. The sale price for neighborhoods Blueste, NWAmes, Greens, Crawfor and ClearCr, it could be a positive sign that this neighborhoods have gained popularity throughout the year and worth investing.<br>

2b. While Timber, IDOTTR and NoRidge sale price is dropping since 2006, it could be a good sign of buying it nothing is wrong with the neighborhood.<br>

2c. For some neighborhoods such as MeadowV, their house price is contantly low , it could be good for those with tight budget.
