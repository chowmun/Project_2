```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
#scatter plot function
def scatter(df, x , y, hue = None, size=7):
    sns.lmplot(x=x, y=y, data=df, fit_reg=False, hue=hue, size=size, scatter_kws={'marker':'o','alpha':0.7,'s':9}, palette='Set2')
```

```python
pd.options.display.max_columns =999 #dispaly the whole grid

train2_csv = '../datasets/train2.csv'
train2 = pd.read_csv(train2_csv)
test2_csv ='../datasets/test2.csv'
test2 = pd.read_csv(test2_csv)
```

I have not concatenate train and test data before data visuallisation, it is to reduce the impact of test dataset toward the analysis.<br>
After the analysis, to save the time from double works and reduce human errors, we will concatenate the train and test data.<br>
To clearly differential both datasets, we will add a column to both datasets. 

```python
train2['from'] = 'train'
test2['from'] = 'test'
df = pd.concat([train2,test2], ignore_index = True)
print( train2.shape, test2.shape, df.shape)
df.tail()
```

## A.Feature Engineering
#### A1. Total Area
Logically, 'SalePrice' is stronly influenced by the amount of square feets in that house. <br>
We will create a column with summation of all square feets.

```python
df['total_area'] = df['Total Bsmt SF'] + df['Gr Liv Area'] 
```

#### A2. Exporting data
##### A2.i.Regrouping Overall Quality

```python
train2['SalePrice'].groupby(train2['Overall Qual']).median().sort_values().plot(kind='bar',color='blue', figsize=(10,5))
```

Regrouping based on the overall quality and price<br>
Poor=1 (for Overall Qual 1,2 and 3) , Average=2 (for Overall Qual 4,5 and 6), Good=3 (for Overall Qual7 and 8) and Excellent=4 (for Overall Qual 9 and 10)

```python
#Simplifications of 'Overall Qual' feature
df['s_overall_q'] = df['Overall Qual'].replace({1 : 1, 2 : 1, 3 : 1, 
                                                   4 : 2, 5 : 2, 6 : 2, 
                                                   7 : 3, 8 : 3, 
                                                   9 : 4, 10 : 4 
                                                   })
```

##### A2.ii.Regrouping Kitchen Qual

```python
#Assign the ordinal variables with appropriate number
df['kitchen_q'] = df['Kitchen Qual'].replace({'Po': 1, 'Fa': 1, 'TA':2, 'Gd':3, 'Ex':4})                                             
```

##### A2.iii.Regrouping Exter Qual

```python
#Assign the ordinal variables with appropriate number
df['exter_q'] = df['Exter Qual'].replace({'Po': 1, 'Fa': 1, 'TA':2, 'Gd':3, 'Ex':4})    
```

## B.One-hot encode categorical variables
Use when column has data without some kind of order or hierarchy.

#### B1. Fireplace
From the EDA, there is a huge difference in price for houses with fireplace and without fireplace but not in the number of fireplace.<br>
We will create a column to check the existing of fireplace in house.

```python
conditions = [(df['Fireplaces']==0), (df['Fireplaces']>0)]
choices = [0, 1]
df['got_fplace'] = np.select( conditions, choices, default='other').astype(int) 
```

#### B2. Foundation

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize= (14,5)) 
sns.countplot(train2['Foundation'], ax=ax1)
sns.boxplot('Foundation', 'SalePrice', data=train2, ax=ax2)
```

We can clearly notice the impact of 'PConc' material on the 'SalePrice'. The rest of the materials have a comparable 'SalePrice' with 'CBlock' whch is the second highest material used. We will create a new column to check is 'PConc' the material for foudation.

```python
conditions = [(df['Foundation'] != 'PConc'), (df['Foundation'] =='PConc')]
choices = [0, 1]
df['pconc'] = np.select( conditions, choices, default='other').astype(int) 
```

#### B3. Garage Finish

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize= (14,5)) 
sns.countplot(train2['Garage Finish'], ax=ax1)
sns.boxplot('Garage Finish', 'SalePrice', data=train2, ax=ax2)
```

From the boxplot, we can see the different in price for houses with 'RFn' and 'Fin' with 'Ubf' and 'None'. <br>
Thus I will regrouping them to check the interior finish of the garage is completed or not. 

```python
df['g_finish'] = df['Garage Finish'].replace({'Unf': 0, 'None': 0, 'RFn':1, 'Fin':1})
```

#### B4. MS Zoning

```python
#Create a dummy for MS Zoning
```

```python
dummy = pd.get_dummies(df['MS Zoning'], drop_first=True, prefix="ms_")
dummy.head()
```

## C.Train/test split data

```python
#Only need to change the name of 'Garaga Area', the fastest way will be add on a new column and use it
df['g_area']=df['Garage Area']
```

```python
selected_features = ['years_of_house', 'reno', 'g_finish', 'pconc', 's_overall_q', 'kitchen_q', 'got_fplace',
                    'g_area', 'exter_q', 'total_area','SalePrice','Id', 'from']
```

```python
final = df[selected_features]
```

Join the dummies created with final

```python
final = final.join([dummy,dummy2])
```

```python
final.head()
```

Separate the original test set and train set from 'final', then remove the column 'from'

```python
test = final[(final['from']=='test')]
test_id = test[['Id']]
test= test.drop(['from', 'SalePrice','Id'], axis=1)
```

```python
#assign X with all the seleceted features, assign y as target or 'SalePrice'
X = final[(final['from']=='train')]
y = X[['SalePrice']]
X = X.drop(['from','SalePrice','Id'], axis=1)
```

```python
X_name = X.columns
test_name = test.columns
```

Train test split for our test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.2)
```

## D.Scale data
Fit StandardScaler into X_train then use it to transform X_train, X_test and test.

```python
ss=StandardScaler()
ss.fit(X_train)
X_train_s = ss.transform(X_train)
X_test_s = ss.transform(X_test)
test_scaled = ss.transform(test) #test data for final verification (no saleprice)
```

```python
#save the scaled datasets
pd.DataFrame(X_train_s, columns=X_name).to_csv('../datasets/X_train.csv')
pd.DataFrame(X_test_s, columns=X_name).to_csv('../datasets/X_test.csv')
pd.DataFrame(test_scaled, columns=test_name).to_csv('../datasets/test_scaled.csv')
y_train.to_csv('../datasets/y_train.csv')
y_test.to_csv('../datasets/y_test.csv')
test_id.to_csv('../datasets/test_id.csv')
```
